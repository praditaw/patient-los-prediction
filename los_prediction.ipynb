### **Latar belakang**

Industri layanan kesehatan, khususnya rumah sakit, secara berkelanjutan menghadapi berbagai tantangan operasional yang kompleks. Beberapa isu yang sering muncul meliputi **peningkatan waktu tunggu pasien di Unit Gawat Darurat (UGD), optimalisasi penggunaan tempat tidur yang belum maksimal, dan tekanan pada biaya operasional termasuk biaya lembur staf.** Di sisi lain, pasien dan keluarga seringkali mengungkapkan kekhawatiran terkait **ketidakpastian durasi perawatan dan mengharapkan proses perencanaan kepulangan yang lebih proaktif dan jelas.**

Menyikapi hal ini, proyek ini bertujuan untuk menganalisis data historis pasien dari berbagai rumah sakit dan mengembangkan model machine learning. Fokus utama adalah **memprediksi rata-rata lama tinggal pasien** (`length_of_stay_avg`), karena metrik ini diidentifikasi sebagai faktor kunci yang mempengaruhi berbagai aspek operasional tersebut di berbagai fasilitas kesehatan.

Model prediksi rata-rata lama tinggal pasien yang akan dibangun ini akan digunakan secara praktis untuk menyelesaikan kasus-kasus operasional berikut di rumah sakit kami:
1. **Menangani kasus antrean pasien di UGD**: dengan prediksi length of stay (LoS) yang lebih baik, tim manajemen RS dapat mengantisipasi ketersediaan tempat tidur lebih cepat, sehingga mengurangi waktu tunggu pasien.
2. **Menangani kasus efisiensi alokasi sumber daya**: perkiraan LoS akan menjadi dasar penjadwalan dinas staf yang lebih sesuai dengan fluktuasi beban kerja, mengurangi potensi lembur dan mengantisipasi kebutuhan logistik (seperti ruang ICU, obat, atau peralatan khusus)
3. **Memperbaiki kasus perencanaan kepulangan (discharge planning)**: informasi prediksi LoS memungkinkan tim medis memulai discharge planning lebih proaktif, memperlancar transisi perawatan dan berpotensi mengurangi readmisi
4. **Menangani masalah komunikasi & kepuasan pasien**: memberikan dasar untuk informasi perkiraan LoS yang lebih reliabel kepada pasien dan keluarga.

Justifikasi mengapa **Length of Stay (LoS) atau Lama Tinggal Pasien** diprediksi dalam satuan **hari** adalah:
1. Banyak penelitian yang bertujuan untuk memprediksi LoS atau menganalisis faktor-faktor yang mempengaruhinya secara implisit atau eksplisit menggunakan "hari" sebagai unit pengukuran. [DOI: 10.1186/s12913-024-11238-y](https://bmchealthservres.biomedcentral.com/articles/10.1186/s12913-024-11238-y)
2. Standar operasional RS: "Hari" adalah unit waktu utama untuk perencanaan tempat tidur, staf, dan keuangan di rumah sakit.
3. Model dilatih menggunakan data historis yang mencatat LoS dalam hari, sehingga model belajar memprediksi dalam unit yang sama.
4. Prediksi harian langsung mendukung keputusan operasional seperti kapan pasien keluar, kebutuhan staf harian, dan perencanaan kepulangan yang lebih proaktif.
5. Hasil dalam "hari" mudah dimengerti dan dikomunikasikan ke semua pihak terkait di rumah sakit.

### **4.3. Hubungan antara Fitur Numerik dengan `length_of_stay_avg`**

Selanjutnya, saya ingin melihat apakah ada tren atau pola ketika membandingkan fitur numerik seperti `admission_count`, `readmission_count`, dll., dengan `length_of_stay_avg`. Saya menggunakan **scatter plot**.

```python
# EDA: Fitur Numerik vs Target
# Fitur numerik yang sudah didefinisikan sebelumnya
nums = ['admission_count', 'readmission_count', 'comorbid_conditions_count', 
        'daily_medication_dosage', 'emergency_visit_count']
target = 'length_of_stay_avg'

# Konfigurasi grid: 3 baris x 2 kolom (total 6 slot)
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(14, 12))
fig.suptitle("Hubungan Fitur Numerik dengan Rata-Rata Lama Tinggal", fontsize=16, y=1.02)

# 'Flatten' array axes agar mudah diiterasi jika n_rows > 1
axes = axes.flatten()

# Plot scatter setiap fitur terhadap target
for i, col in enumerate(nums):
    sns.scatterplot(x=df[col], y=df[target], alpha=0.4,
        ax=axes[i],color='#4073FF')
    axes[i].set_title(f'{col} vs {target}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel(target)
    axes[i].grid(True)

# Mengosongkan subplot terakhir jika jumlah fitur ganjil (slot ke-6)
if len(nums) < len(axes):
    axes[-1].axis('off')

plt.tight_layout()
plt.show()

# Untuk ringkasan korelasi Pearson dari fitur numerik
nums_corr = nums + [target]
corr= df[nums_corr].corr()
print('Koefisien korelasi antara fitur numerik dan length_of_stay_avg:')
corr
```

Hubungan Fitur Numerik dengan `length_of_stay_avg`:

### **4.4. Analisis Hubungan Semua Fitur dengan Target**

Setelah melihat bahwa korelasi linear (Pearson) antara fitur numerik dan `length_of_stay_avg` sangat lemah, saya menggunakan koefisien korelasi Phik yang dapat menangkap berbagai jenis hubungan, termasuk yang non-linear. Skala korelasi berkisar dari **0 (tidak berkorelasi)** hingga **1 (korelasi sangat kuat)**.


```python
# Menghitung korelasi menggunakan phik
# Mendefinisikan kolom-kolom yang akan dimasukkan dalam analisis Phik
nums = ['admission_count', 'readmission_count', 
        'comorbid_conditions_count', 'daily_medication_dosage', 
        'emergency_visit_count']

# Kolom kategorikal asli
cats = ['hospital_name', 'condition_type', 'patient_age_group', 
        'patient_gender', 'severity_level', 'seasonal_indicator', 
        'primary_diagnosis_code']

# Ekstraksi Fitur dari 'admission_date'
# Mengambil tahun, bulan, hari dalam seminggu
df['admission_year'] = df['admission_date'].dt.year
df['admission_month'] = df['admission_date'].dt.month_name() # Nama bulan (misal: January)
df['admission_day_of_week'] = df['admission_date'].dt.day_name() # Nama hari (misal: Monday)

# Fitur waktu baru (ini akan diperlakukan sebagai kategorikal oleh Phik)
time = ['admission_year', 'admission_month', 
        'admission_day_of_week']

# Kolom target
target = 'length_of_stay_avg'

# Menggabungkan semua kolom yang relevan
all_for_phik = nums + cats + time + [target]
df_phik = df[all_for_phik].copy()

# Menghitung matriks korelasi Phik
phik_corr = df_phik.phik_matrix(interval_cols=nums + [target])

# Visualisasi heatmap korelasi
plt.figure(figsize=(18, 15)) # Sesuaikan ukuran jika perlu
sns.heatmap(phik_corr, 
            annot=True, # Menampilkan angka korelasi
            fmt=".2f",  # Format angka 2 desimal
            cmap="YlGn", # Skema warna (coolwarm_r agar merah = asosiasi kuat)
            linewidths=.5,
            vmin=0, vmax=1) # Phik berkisar 0-1
plt.title('Heatmap Korelasi Phik Antar Semua Fitur (Termasuk Target)', fontsize=15)
plt.tight_layout()
plt.show()

# Menampilkan korelasi Phik semua fitur dengan variabel target secara spesifik
print("Korelasi Phik semua fitur dengan 'length_of_stay_avg':")
phik_with_target = phik_corr['length_of_stay_avg'].sort_values(ascending=False)
print(phik_with_target)
```

Heatmap secara keseluruhan didominasi warna kuning muda, yang menunjukkan nilai Phik sangat rendah (mendekati 0) di hampir semua pasangan fitur. Ini berarti, **bahkan ketika memperhitungkan hubungan non-linear, hubungan antar fitur (termasuk dengan `length_of_stay_avg`) umumnya sangat lemah**.

Fitur seperti `severity_level`, `patient_gender`, dan `daily_medication_dosage` menunjukkan nilai Phik tertinggi dengan `length_of_stay_avg`, namun nilainya masih sangat kecil (sekitar 0.01-0.02). Sejumlah besar fitur lain (seperti `admission_count`, `hospital_name`, `condition_type`, `primary_diagnosis_code`, dll.) memiliki korelasi Phik praktis nol dengan `length_of_stay_avg`.

Analisis Phik mengonfirmasi dan memperkuat temuan sebelumnya bahwa fitur-fitur pada dataset ini, bahkan ketika dianalisis untuk hubungan non-linear, menunjukkan hubungan individual yang sangat lemah dengan `length_of_stay_avg`. **Tidak ada satu pun fitur yang menonjol sebagai prediktor kuat secara tunggal**.

### **5.7. Outliers Handling**

```python
# Mengecek bentuk distribusi skewness
all_num_cols = list(X_train.select_dtypes(include=np.number).columns)
column_to_exclude = 'admission_year' # Kolom ini akan diperlakukan sebagai kategorikal
num_cols = [col for col in all_num_cols if col != column_to_exclude]

# Simpan hasil ke dalam list
listItem = []

for col in num_cols:
    # Menghitung skewness
    skew_val = round(X_train[col].skew(), 1)
    distribusi = 'normal' if -0.5 <= skew_val <= 0.5 else 'skewed'

    # Menghitung IQR
    Q1 = X_train[col].quantile(0.25)
    Q3 = X_train[col].quantile(0.75)
    IQR = Q3 - Q1

    # Batas deteksi outlier
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Menghitung persentase outlier
    outlier_count = X_train[(X_train[col] < lower_bound) | (X_train[col] > upper_bound)].shape[0]
    total_count = X_train[col].shape[0]
    outlier_percentage = round((outlier_count / total_count) * 100, 2)

    # Menambahkan ke list
    listItem.append([col, skew_val, distribusi, outlier_percentage])

# Membuat DataFrame hasil
skewness = pd.DataFrame(listItem, columns=['kolom', 'nilai skewness', 'distribusi', 'persentase_outlier (%)'])
skewness
```

Analisis ini menunjukkan bahwa sebagian besar fitur numerik (`admission_count`, `readmission_count`, `comorbid_conditions_count`, `emergency_visit_count`) memiliki **distribusi yang miring ke kanan (positively skewed)**. Fitur `daily_medication_dosage` adalah satu-satunya yang menunjukkan distribusi normal.
Meskipun beberapa fitur miring, **persentase outlier yang terdeteksi oleh metode IQR untuk semua fitur numerik ini relatif rendah (semuanya di bawah 1.5%)**.

Karena persentase outlier tidak sangat tinggi, saya bisa mempertimbangkan metode capping (Winsorizing) untuk menangani nilai-nilai ekstrem ini tanpa kehilangan banyak data.

**Outlier Handling - menggunakan `Winsorizer` dari library `feature_engine` untuk melakukan capping outlier.**

```python
# Daftar fitur yang akan dicapping
features_to_winsorize = ['admission_count',
                         'readmission_count',
                         'comorbid_conditions_count',
                         'emergency_visit_count']
# Capping
# Membuat objek Winsorizer dengan metode IQR dan batas atas bawah (1.5 IQR)
winsor = Winsorizer(
    capping_method='iqr',
    tail='both',            # Melakukan capping pada kedua ekor (bawah dan atas)    
    fold=1.5,                   
    variables=features_to_winsorize)

# Fit transformer pada data latih dan transform data latih
X_train_capped = winsor.fit_transform(X_train)
# Transform data uji menggunakan parameter yang sama (dipelajari dari data latih)
X_test_capped = winsor.transform(X_test)
```

Untuk melihat dampak dari proses capping ini, saya membuat fungsi plots yang menampilkan histogram dan boxplot. Saya kemudian menjalankan fungsi ini untuk setiap fitur, membandingkan visualisasi sebelum (`X_train`) dan sesudah (`X_train_capped`)

```python
# Membuat visualisasi histogram dan boxplot
def plots(X_train_capped, variables):
    plt.figure(figsize=(16, 4))

    # Menampilkan Histogram
    plt.subplot(1, 2, 1)
    sns.histplot(X_train_capped[variables], bins=30, color='crimson')
    plt.title('Histogram')

    # Menampilkan Boxplot
    plt.subplot(1, 2, 2)
    sns.boxplot(y=X_train_capped[variables], color='crimson')
    plt.title('Boxplot')

    # Menampilkan visualisasi
    plt.show()

# Melihat distribusi data sebelum dan setelah dilakukan capping
for col in features_to_winsorize:
  print(f'{col} - Sebelum dilakukan Capping')
  plots(X_train, col)

  print(f'\n{col} - Sesudah dilakukan Capping')
  plots(X_train_capped, col)
```

Visualisasi boxplot mengonfirmasi bahwa **proses capping dengan Winsorizer ini telah berhasil mengurangi dampak dari nilai-nilai ekstrem (outlier) pada fitur-fitur numerik yang dipilih**.

Misalnya pada `readmission_count`, sebelum capping, boxplotnya menunjukkan ada beberapa kasus dengan jumlah readmisi yang cukup tinggi, jauh di atas mayoritas data yang bernilai 0 atau 1. Setelah capping, titik-titik outlier ini sudah tidak ada lagi; nilai-nilai ekstrem tersebut kini sudah disesuaikan menjadi nilai batas atas yang dihitung dari IQR data latih. Histogramnya juga akan menunjukkan frekuensi yang lebih terkumpul di nilai-nilai yang lebih rendah setelah nilai-nilai tinggi 'ditarik' masuk.

```python
# Membuat fungsi untuk menghitung Phi-K correlation
def compute_phik_correlation(dataframe, columns, target):

    # Mengambil subset dari DataFrame
    subset = dataframe[columns]
    # Menghitung matriks Phi-K
    correlation_matrix = subset.phik_matrix()
    # Mengembalikan kolom target dari matriks
    return correlation_matrix[target]

# Definisikan list kolom untuk setiap subset
columns = ['hospital_name', 'admission_count', 'condition_type',
       'patient_age_group', 'patient_gender', 'readmission_count',
       'severity_level', 'seasonal_indicator', 'comorbid_conditions_count',
       'primary_diagnosis_code', 'daily_medication_dosage',
       'emergency_visit_count', 'admission_year', 'admission_month',
       'admission_day_of_week', 'length_of_stay_avg']
target = ['length_of_stay_avg']

# Compute Phi-K correlation for each set of columns and print
correlation = compute_phik_correlation(train_data, columns, target)

# Menampilkan hasil
print(correlation)
```

### **5.10. Column Transformer**

Selanjutnya, saya menyiapkan alat untuk pra-pemrosesan, yaitu `ColumnTransformer` yang saya beri nama `preprocessor`.
* Pada kolom numerik (`numerical_cols_selected`) akan diterapkan `StandardScaler` untuk mengubah skala nilainya (rata-rata 0, standar deviasi 1).
* Pada kolom kategorikal ordinal (`ordinal_categorical_cols` seperti `patient_age_group`, `severity_level`)akan diterapkan `OrdinalEncoder`. Ini akan mengubah kategori seperti 'Mild', 'Moderate', 'Severe' menjadi angka berurutan (misalnya 0, 1, 2) sesuai dengan urutan (`order_patient_age_group`, `order_severity_level`) yang sudah saya tentukan. Parameter `handle_unknown`='use_encoded_value', `unknown_value`=-1 digunakan untuk menangani nilai tak dikenal di data uji.
* Pada kolom kategorikal nominal (`nominal_categorical_cols` seperti `condition_type`, `patient_gender` dll)akan diterapkan `OneHotEncoder`. Ini akan membuat kolom biner baru untuk setiap kategori, dengan drop='first' untuk menghindari multikolinearitas dan handle_unknown='ignore' untuk kategori baru di data uji.

```python
# Membuat objek preprocessor dengan ColumnTransformer
order_patient_age_group = ['0-17', '18-45', '46-65', '66+'] 
order_severity_level = ['Mild', 'Moderate', 'Severe'] 

preprocessor= ColumnTransformer(
    transformers=[
        ('scaler', StandardScaler(), numerical_cols_selected),
        ('ord_encode', OrdinalEncoder(categories=[order_patient_age_group, order_severity_level], 
                               handle_unknown='use_encoded_value', unknown_value=-1), # Mengatasi unknown di test
                         ordinal_categorical_cols),
        ('cat_ohe', OneHotEncoder(handle_unknown='ignore', drop='first', sparse_output=False), 
                          nominal_categorical_cols)
    ],
    remainder='passthrough'
)
```

Tahap ini bertujuan untuk mengevaluasi seberapa baik performa masing-masing model yang sudah dilatih, baik pada data latih maupun pada data uji.

Dalam mengevaluasi dan mengoptimalkan model prediksi length_of_stay_avg ini, saya akan menggunakan Mean Absolute Error (MAE) sebagai metrik evaluasi utama. Sementara metrik $R^2$ (R-squared) tetap akan dilaporkan sebagai informasi pelengkap, fokus pada MAE didasarkan pada beberapa pertimbangan praktis dan relevansinya dengan tujuan proyek untuk Tim Manajemen Rumah Sakit.

**Justifikasi pemilihan MAE sebagai metrik evaluasi:**
* **Interpretasi Langsung (bisa dijelaskan langsung pada User**): MAE secara langsung mengukur rata-rata besarnya kesalahan prediksi dalam satuan asli target, yaitu 'hari'. Jika model memiliki MAE sebesar 1.1 hari, artinya secara rata-rata, prediksi lama tinggal pasien oleh model akan meleset sekitar 1.1 hari dari kenyataan. Besaran kesalahan ini sangat mudah dipahami oleh semua pihak, termasuk Tim Manajemen dan staf klinis, serta langsung dapat dikaitkan dengan dampak operasional seperti perencanaan ketersediaan tempat tidur, penjadwalan staf, dan estimasi kebutuhan sumber daya.
* **Fokus pada kesalahan bbsolut rata-rata**: dalam konteks operasional rumah sakit, seringkali yang paling penting adalah seberapa jauh prediksi meleset secara rata-rata, tanpa mempedulikan apakah prediksi tersebut lebih tinggi atau lebih rendah dari nilai aktual. MAE secara adil merepresentasikan besaran kesalahan ini.

* **Ketahanan terhadap outlier dalam kesalahan prediksi**: MAE kurang sensitif terhadap beberapa prediksi yang memiliki kesalahan sangat besar (outlier dalam error). Dalam kasus LoS, mungkin ada beberapa pasien dengan kondisi tak terduga yang membuat LoS-nya sangat berbeda. MAE akan memberikan gambaran performa yang lebih stabil untuk mayoritas pasien.

Beberapa referensi yang memprediksi LoS dengan metriks evaluasinya adalah MAE:

*"The eXtreme Gradient Boosting Regression (XGBR) model had the lowest prediction error, both in terms of RMSPE (11.00 days) and MAE (7.52 days)."* [Zeleke et al.](https://www.sciencedirect.com/science/article/pii/S2352914824000558)

*"A LoS study for patients with sepsis [28] compared six ML regression models, namely LR, RFR, KNN, NN, XGBR, and light GBR, and found that light GBM showed the best model having an MAE of 4 days."* [Chen et all](https://ieeexplore.ieee.org/document/9764052)

*"They evaluated several machine learning methods and found that a neural network (NN) algorithm had the lowest mean absolute error (MAE) of 2.23 days for predicting LoS."* [Kolchun et all](https://thejns.org/focus/view/journals/neurosurg-focus/52/4/article-pE12.xml)

# **9. Modeling - Hyperparameter Tuning**
---

Meskipun `Gradient Boosting Regressor` menunjukkan MAE CV terendah di antara model baseline, performa $R^2$ score-nya masih belum memuaskan. Saya akan mencoba melakukan hyperparameter tuning untuk model ini guna mencari konfigurasi parameter terbaik yang mungkin bisa meningkatkan kinerjanya secara keseluruhan.

Tujuan dari *Hyperparameter Tuning* adalah untuk mencari kombinasi pengaturan (hyperparameter) terbaik untuk model yang dipilih, sehingga diharapkan dapat meningkatkan performa prediksinya (menurunkan MAE dan menaikkan $R^2$).

Hasil evaluasi model-model baseline dan upaya tuning awal menunjukkan bahwa skor $R^2$ secara konsisten sangat rendah (negatif atau mendekati nol). **Ini mengindikasikan bahwa dengan fitur-fitur yang tersedia saat ini, model kesulitan untuk menjelaskan sebagian besar variabilitas dalam `length_of_stay_avg`**. Dalam situasi seperti ini, optimasi pada $R^2$ mungkin kurang produktif. Sebaliknya, fokus pada minimalisasi MAE menjadi lebih bisa dilakukan. Tujuannya adalah untuk menghasilkan model yang, meskipun mungkin tidak bisa menjelaskan mengapa LoS bervariasi secara mendalam, setidaknya dapat memberikan estimasi dengan tingkat kesalahan harian rata-rata yang paling minimal dan dapat diterima secara operasional.

### **9.1. Hyperparameter Tuning - Percobaan ke-1**

#### **Model Definition**

```python
# Pipeline Gradient Boosting Regressor
pipeline_gb_tuning = Pipeline([
    ('preprocessor', preprocessor),
    ('regressor', GradientBoostingRegressor(random_state=42))
])

# Mendefinisikan Random Hyperparameter untuk GradientBoostingRegressor
param_distributions_gb = {
    'regressor__n_estimators': [100, 200, 300, 400, 500],  # Jumlah pohon
    'regressor__learning_rate': [0.01, 0.025, 0.05, 0.1], # Laju pembelajaran
    'regressor__max_depth': [3, 4, 5, 6],                 # Kedalaman maksimum setiap pohon
    'regressor__min_samples_split': [2, 5, 10, 15],      # Jumlah minimum sampel untuk membagi node
    'regressor__min_samples_leaf': [1, 2, 4, 6],         # Jumlah minimum sampel pada leaf node
    'regressor__subsample': [0.7, 0.8, 0.9, 1.0],         # Fraksi sampel
    'regressor__max_features': ['sqrt', 'log2', None] 
}
```

```python
# Menyiapkan KFold
kfold = KFold(n_splits=5, shuffle=True, random_state=42)
# Inisialisasi RandomizedSearchCV
random_search_gb = RandomizedSearchCV(
    estimator=pipeline_gb_tuning, 
    param_distributions=param_distributions_gb, 
    n_iter=50,  
    cv=kfold,       
    scoring='neg_mean_absolute_error', 
    n_jobs=-1,
    random_state=42, # Untuk reproduktifitas
    verbose=1 
)
```

```python
# Pencarian Hyperparameter Terbaik pada Data Latih
random_search_gb.fit(X_train_selected, y_train)
print("RandomizedSearchCV selesai.")
```

Saya menggunakan `RandomizedSearchCV` untuk mencoba 50 kombinasi *hyperparameter* yang berbeda untuk `GradientBoostingRegressor`. Setiap kombinasi dievaluasi menggunakan 5-fold cross-validation pada data latih.

```python
# Menampilkan Hyperparameter Terbaik dan Skor MAE Terbaik dari Cross-Validation
print("Hyperparameter terbaik yang ditemukan:")
print(random_search_gb.best_params_)

best_mae_cv_gb_tuned = -random_search_gb.best_score_ # Ubah kembali ke MAE positif
print(f"\nMAE Cross-Validation terbaik dari RandomizedSearchCV: {best_mae_cv_gb_tuned}")

# Mendapatkan Model Terbaik (Pipeline Terbaik)
best_pipeline_gb = random_search_gb.best_estimator_
```

Kombinasi terbaik yang ditemukan adalah:
* `subsample`: 0.7 (menggunakan 70% data untuk setiap pohon)
* `n_estimators`: 100 (jumlah pohon)
* `min_samples_split`: 15 (jumlah sampel minimum untuk melakukan split)
* `min_samples_leaf`: 4 (jumlah sampel minimum di leaf node)
* `max_features`: None (mempertimbangkan semua fitur saat split)
* `max_depth`: 3 (kedalaman pohon yang relatif dangkal)
* `learning_rate`: 0.01 (laju pembelajaran yang kecil)

Rata-rata MAE terendah yang dicapai selama proses cross-validation dengan kombinasi parameter terbaik adalah 1.2046 hari, yang berarti pada data latih (di berbagai fold), model ini secara rata-rata salah memprediksi lama tinggal pasien sekitar 1.205 hari.

#### **Model Evaluation**

```python
# 7. Evaluasi Model Terbaik pada Data Uji
print("Evaluasi Model Gradient Boosting Regressor (Tuned by RandomizedSearch) pada Data Uji:")
y_pred_test_gb_tuned = best_pipeline_gb.predict(X_test_selected)

mae_test_gb_random_tuned = mean_absolute_error(y_test, y_pred_test_gb_tuned)
r2_test_gb_random_tuned = r2_score(y_test, y_pred_test_gb_tuned)

print(f"MAE pada data uji (Tuned by RandomizedSearch): {mae_test_gb_random_tuned}")
print(f"R2 pada data uji (Tuned by RandomizedSearch): {r2_test_gb_random_tuned}")
```

* Saat model dengan parameter terbaik ini diuji pada data baru (`X_test_selected`), **MAE yang dihasilkan adalah sekitar 1.188 hari**. Nilai ini sedikit lebih baik daripada MAE rata-rata dari cross-validation, yang merupakan hal yang positif dan menunjukkan model tidak mengalami degradasi performa yang signifikan pada data uji.
* Meskipun MAE Test cukup dekat dengan MAE CV, skor $R^2$ pada data uji masih negatif dan sangat mendekati nol. Ini berarti bahkan setelah tuning dengan fokus pada MAE, model ini masih belum mampu menjelaskan variasi dalam `length_of_stay_avg` lebih baik daripada kalau kita asal tebak pakai nilai rata-rata saja.

Hasil ini menunjukkan bahwa meskipun bisa sedikit menekan MAE, **keterbatasan utama kemungkinan masih pada daya prediktif fitur-fitur yang ada** (seperti dibuktikan dalam perhitungan nilai korelasi phik). Kombinasi parameter yang terpilih juga mengindikasikan bahwa model cenderung 'berhati-hati' untuk tidak overfitting pada data yang mungkin memiliki sinyal lemah atau noise.

### **9.2. Hyperparameter Tuning - Percobaan ke-2**

#### **Model Definition**

Saya melakukan upaya tuning kedua untuk `GradientBoostingRegressor` dengan rentang parameter yang berbeda dan `n_iter` yang lebih banyak (75), masih dengan fokus pada MAE sebagai metrik utama.

```python
# Pipeline Gradient Boosting Regressor
pipeline_gb_tuning_attempt2 = Pipeline([
    ('preprocessor', preprocessor),
    ('regressor', GradientBoostingRegressor(random_state=42))
])
# Mendefinisikan Distribusi/Rentang Hyperparameter BARU (Upaya ke-2)
param_distributions_gb_attempt2 = {
    'regressor__n_estimators': [100, 200, 300, 500, 700],  # Rentang lebih luas, termasuk lebih banyak pohon
    'regressor__learning_rate': [0.005, 0.01, 0.02, 0.05, 0.1], # Mencoba learning rate yang sangat kecil dan sedang
    'regressor__max_depth': [3, 4, 5, 6, 8],             # Memberi kesempatan pohon lebih dalam lagi
    'regressor__min_samples_split': [10, 20, 30, 40],    # Meningkatkan nilai untuk regularisasi lebih kuat
    'regressor__min_samples_leaf': [5, 10, 15, 20],       # Meningkatkan nilai untuk regularisasi lebih kuat
    'regressor__subsample': [0.6, 0.7, 0.8, 0.9],        # Variasi subsample
    'regressor__max_features': ['sqrt', 'log2', 0.7, None], # Variasi max_features, None berarti semua fitur
    'regressor__loss': ['squared_error', 'absolute_error', 'huber'] # Coba fungsi loss berbeda
}
```

```python
# Inisialisasi RandomizedSearchCV
# Meningkatkan n_iter untuk eksplorasi yang lebih baik dari ruang parameter yang lebih beragam ini
random_search_gb_attempt2 = RandomizedSearchCV(
    estimator=pipeline_gb_tuning_attempt2, 
    param_distributions=param_distributions_gb_attempt2, 
    n_iter=75,  
    cv=kfold,       
    scoring='neg_mean_absolute_error',
    n_jobs=-1,
    random_state=42,
    verbose=1 
)
```

```python
# Pencarian Hyperparameter Terbaik pada Data Latih
random_search_gb_attempt2.fit(X_train_selected, y_train) 
print("RandomizedSearchCV (Upaya ke-2) selesai.")
```

Pada percobaan kedua ini, `RandomizedSearchCV` mencoba 75 kombinasi hyperparameter yang berbeda, masing-masing dievaluasi dengan 5-fold cross-validation (total 375 model dilatih).

```python
# Menampilkan Hyperparameter Terbaik dan Skor MAE Terbaik dari Cross-Validation
print("Hyperparameter terbaik yang ditemukan (Upaya ke-2):")
print(random_search_gb_attempt2.best_params_)

best_mae_cv_gb_tuned_attempt2 = -random_search_gb_attempt2.best_score_ # Ubah kembali ke MAE positif
# Menampilkan MAE Cross-Validation Terbaik dari RandomizedSearchCV (Upaya ke-2)
print(f"\nMAE Cross-Validation terbaik dari RandomizedSearchCV (Upaya ke-2): {best_mae_cv_gb_tuned_attempt2}")

# Model Terbaik (Pipeline Terbaik)
best_pipeline_gb_attempt2 = random_search_gb_attempt2.best_estimator_
```

Kombinasi terbaik yang ditemukan pada upaya ke-2 adalah:
- `subsample`: 0.9
- `n_estimators`: 100
- `min_samples_split`: 20 (lebih tinggi, untuk regularisasi)
- `min_samples_leaf`: 5 (lebih tinggi, untuk regularisasi)
- `max_features`: 0.7 (menggunakan 70% fitur per split)
- `max_depth`: 5 (sedikit lebih dalam dari upaya pertama)
- `loss`: 'huber' (fungsi loss yang berbeda, robust terhadap outlier)
- `learning_rate`: 0.005 (laju pembelajaran yang sangat kecil)

Rata-rata MAE terendah dari cross-validation adalah sekitar 1.205 hari. Nilai ini hampir sama dengan hasil MAE CV dari upaya tuning pertama (1.2046 hari). Ini menunjukkan bahwa meskipun mencoba banyak kombinasi parameter yang berbeda, performa terbaik yang bisa dicapai pada data latih (via CV) dari sisi MAE tidak banyak berubah.

#### **Model Evaluation**

    * Mempertimbangkan pendekatan pemodelan alternatif, misalnya jika regresi untuk nilai LoS eksak terbukti sangat sulit, evaluasi apakah mengubah masalah menjadi klasifikasi kategori LoS (misalnya, Pendek: 1-3 hari, Sedang: 4-7 hari, Panjang: >7 hari) bisa lebih berhasil dan tetap bermanfaat secara operasional.
